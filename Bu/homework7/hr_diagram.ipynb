```python
import numpy as np
import matplotlib.pyplot as plt
# i manually pulled the files in the terminal 
#(saving homework7 then going back to main and pulling)
# im not sure if its the way to do it
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt",
skiprows=1)
```

```python
print(data.shape) 
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
print(magnitude.min(), magnitude.max())
print(temperature.min(), temperature.max())
```

```python
#white dwarfs are the brightest, supergiants are the dimmest
```

```python
log_mag = np.log(magnitude)
log_temp = np.log(temperature)
```

```python
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(log_temp, log_mag, s = 2)
```

```python
plt.scatter(log_temp, log_mag, s = 2)
ax = plt.gca()
ax.invert_xaxis()
ax.invert_yaxis()

plt.show()
```

```python
brightest = np.argmin(magnitude)
dimmest = np.argmax(magnitude)
hottest = np.argmax(temperature)
coldest = np.argmin(temperature)

plt.scatter(log_temp[brightest], log_mag[brightest], marker="*", s=300, color="black",  label="Brightest")
plt.scatter(log_temp[dimmest],   log_mag[dimmest],   marker="*", s=300, color="purple", label="Dimmest")
plt.scatter(log_temp[hottest],   log_mag[hottest],   marker="*", s=300, color="green",  label="Hottest")
plt.scatter(log_temp[coldest],   log_mag[coldest],   marker="*", s=300, color="orange", label="Coldest")

plt.title("HertzsprungRussell Diagram", fontsize=25)
plt.xlabel("Temperature (log K)", fontsize=25)
plt.ylabel("Magnitude (log scale)", fontsize=25)
plt.legend(fontsize=25)
```

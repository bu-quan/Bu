```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2*np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 5))
```

```python
for A, w in zip(amplitude, frequency):
    y1 = make_sine_wave(x, A, 1)
    axes[0].plot(x, y1, label=f'A={A}, w=1')
    y2 = make_sine_wave(x, 1, w)
    axes[1].plot(x, y2, label=f'A=1, w={w}')
    y3 = make_sine_wave(x, A, w)
    axes[2].plot(x, y3, label=f'A={A}, w={w}')
```

```python
axes[0].set_title('Amplitude')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()

axes[1].set_title('Frequency')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].legend()

axes[2].set_title('Both')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].legend()
```

```python
plt.show()
```
